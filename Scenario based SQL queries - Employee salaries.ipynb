### 17. What is the average salary for each job title, and what is the total number of jobs available for each job title?

### 18. What are the top 5 locations with the highest total salaries, and what is the total number of jobs available for each location?

```python
%sql SELECT company_location, SUM(salary) AS total_salary, COUNT(*) AS num_jobs FROM data_salaries GROUP BY company_location ORDER BY total_salary DESC LIMIT 5;
```

### 19. What is the average salary,  percentage of jobs for each job title in each location?

```python
%sql SELECT job_title, company_location, AVG(salary) AS average_salary, (COUNT(*) * 100 / (SELECT COUNT(*) FROM data_salaries WHERE company_location = j.company_location)) AS job_percentage FROM data_salaries j GROUP BY job_title, company_location;
```

### 20. What is the average salary for each job title, and what is the percentage of jobs for each job title in the dataset?

```python
%sql SELECT job_title, count(job_title), round(AVG(salary)) AS average_salary, (COUNT(*) * 100 / (SELECT COUNT(*) FROM data_salaries)) AS 'job_percentage' FROM data_salaries GROUP BY job_title order by job_percentage desc;
```

### 22. What are the top job titles with the highest salaries in each location?

 ### 24. What are the top 5 locations with the highest average salaries, and what is the name of the company that offers the highest salary for each city?

```python
%sql SELECT company_location, AVG(salary) AS average_salary, company_name FROM data_salaries INNER JOIN data_company ON data_salaries.id = data_company.id GROUP BY company_location ORDER BY average_salary DESC LIMIT 5;
```

### 25. What is the total number of jobs available for each job title in each location, and what is the rank of each job title within each location based on the total number of jobs?

```python
%sql SELECT job_title, company_location, COUNT(*) AS num_jobs, RANK() OVER (PARTITION BY company_location ORDER BY COUNT(*) DESC) AS job_rank FROM data_salaries GROUP BY job_title, company_location;
```

### 26. Find out the company name, company size ?

```python
%sql SELECT company_name, company_size, AVG(salary) FROM data_salaries INNER JOIN data_company ON data_salaries.id = data_company.id GROUP BY company_name order by AVG(salary) desc;
```
